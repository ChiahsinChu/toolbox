# `use_style`

```python
import matplotlib.pyplot as plt
import numpy as np

from intdielec import plot
```

```python
plt.plot(np.arange(10), np.arange(10), label="test")
plt.xlabel("test")
plt.legend()

plt.show()
```

```python
plot.use_style("pre")

plt.plot(np.arange(10), np.arange(10), label="test")
plt.xlabel("test")
plt.legend()

plt.show()
```

```python
plot.use_style("pub")

plt.plot(np.arange(10), np.arange(10), label="test")
plt.xlabel("test")
plt.legend()

plt.show()
```
